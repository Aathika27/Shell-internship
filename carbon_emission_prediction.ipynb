<h3>RENAMING THE COLUMN</h3>

```python
df.rename(columns=lambda x: x.strip().replace(" ", "_").lower(), inplace=True)
```

<h3>Handle Missing Values</h3>

```python
num_cols = df.select_dtypes(include=['float64','int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
df.isnull().sum()
```

<h3>REMOVING DUPLICATES<</h3>

```python
df = df.drop_duplicates()
```

```python
# Convert food_product to string
df["food_product"] = df["food_product"].astype(str)

# Convert all other columns to float
numeric_cols = df.columns.drop("food_product")
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
print(df.dtypes)
```

<h3>DATA EXPLORATION</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(6,4))
    df[column].hist(bins=20, edgecolor='black')
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
if "Co2_emissions" in df.columns or "CO2_emissions" in df.columns:
    col_name = "Co2_emissions" if "Co2_emissions" in df.columns else "CO2_emissions"
    top10 = df.nlargest(10, col_name)[["Food product", col_name]]
    plt.figure(figsize=(10,6))
    sns.barplot(x=col_name, y="Food product", data=top10, palette="viridis")
    plt.title("Top 10 Foods with Highest CO2 Emissions")
    plt.show()
```
